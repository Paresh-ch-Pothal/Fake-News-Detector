```python
import re
```

```python
def clean_text(text):
    if not isinstance(text, str):  # handle potential NaNs or non-string types
        return ""
    # Remove all non-alphanumeric characters except spaces
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)
    # Convert to lowercase (optional but recommended)
    text = text.lower()
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    return text.strip()
```

```python
import nltk
from nltk.corpus import stopwords
import re

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
def clean_and_remove_stopwords(text):
    if not isinstance(text, str):
        return ""

    # Remove special characters and digits
    text = re.sub(r'[^A-Za-z\s]', '', text)

    # Lowercase
    text = text.lower()

    # Tokenize
    words = text.split()

    # Remove stopwords
    filtered_words = [word for word in words if word not in stop_words]

    return ' '.join(filtered_words)
```

```python
df['text'] = df['text'].apply(clean_and_remove_stopwords)
```

Splitting the text to train and test

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
ytest
```

Doing the tokenization part

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(xtrain)
```

```python
tokenizer.word_index
```

```python
xtrain_sequences = tokenizer.texts_to_sequences(xtrain)
xtest_sequences = tokenizer.texts_to_sequences(xtest)
```

```python
# giving padding
maxlen = 100
xtrain_padded = pad_sequences(xtrain_sequences,maxlen = maxlen)
xtest_padded = pad_sequences(xtest_sequences,maxlen=maxlen)
```

Building the model

```python
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Dense , Flatten , Dropout , Embedding , LSTM , Input
```

```python
input = Input(shape=(maxlen,))
vocab_size = len(tokenizer.word_index) + 1
x = Embedding(input_dim=vocab_size,output_dim = 100)(input)
x = Dropout(0.5)(x)
x = LSTM(150)(x)
x = Dropout(0.5)(x)
x = Dense(128,activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(2,activation='softmax')(x)

model = Model(input,x)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder = LabelEncoder()
ytrain_encoded = labelEncoder.fit_transform(ytrain)
ytest_encoded = labelEncoder.transform(ytest)
```

```python

ytrain
```

```python
ytrain_cat = tf.keras.utils.to_categorical(ytrain_encoded)
ytest_cat = tf.keras.utils.to_categorical(ytest_encoded)
```

```python
ytrain_cat
```

```python
history = model.fit(xtrain_padded,ytrain_cat,epochs=2,batch_size=32,validation_data=(xtest_padded,ytest_cat))
```

#Prediction

```python
def func(text):
  cleaned_text = clean_and_remove_stopwords(text)
  input_text = tokenizer.texts_to_sequences([cleaned_text])
  input_text_pad = pad_sequences(input_text,maxlen=maxlen)
  res = model.predict(input_text_pad)
  zero = res[0][0]
  one = res[0][1]
  if zero > one:
    print("False News")
  else:
    print("True News")
```

# Evaluation Of Model


```python
from sklearn.metrics import accuracy_score

# Prediction
y_pred = model.predict(xtest_padded)
y_pred_labels = y_pred.argmax(axis=1)  # Convert softmax to class labels

# Convert one-hot encoded ytest_cat to label format
y_true = ytest_cat.argmax(axis=1)

# Now compute accuracy
accuracy = accuracy_score(y_true, y_pred_labels)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

y_true = ytest_cat.argmax(axis=1)
cm = confusion_matrix(y_true, y_pred_labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Fake", "True"])
disp.plot(cmap="Blues")
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_true, y_pred_labels, target_names=["Fake", "True"]))
```

```python
from sklearn.metrics import roc_curve, auc

# For binary classification, you need probability for class 1
y_scores = y_pred[:, 1]
fpr, tpr, thresholds = roc_curve(y_true, y_scores)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```
